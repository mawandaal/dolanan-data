## Membuat Line Plot dari Data Tabular untuk Kategori

```python
list(df_tab_kat)
```

```python
sns.lineplot(x = pd.Series(pd.to_datetime(df_tab_kat.index)).dt.day, y = df_tab_kat['Bumbu'].values, label = 'Bumbu')
sns.lineplot(x = pd.Series(pd.to_datetime(df_tab_kat.index)).dt.day, y = df_tab_kat['Karbohidrat'].values, label = 'Karbohidrat')
sns.lineplot(x = pd.Series(pd.to_datetime(df_tab_kat.index)).dt.day, y = df_tab_kat['Minuman'].values, label = 'Minuman')
sns.lineplot(x = pd.Series(pd.to_datetime(df_tab_kat.index)).dt.day, y = df_tab_kat['Minyak Memasak'].values, label = 'Minyak Memasak')
sns.lineplot(x = pd.Series(pd.to_datetime(df_tab_kat.index)).dt.day, y = df_tab_kat['Protein'].values, label = 'Protein')
plt.ylim((0,75000))
plt.legend()
plt.show();
```
